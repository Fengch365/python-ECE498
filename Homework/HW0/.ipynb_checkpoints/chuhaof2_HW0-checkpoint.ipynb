\textbf{Problem 4 (Exponential Distributions and Poisson Distributions)}\\ \ \\
\text{(a) (i)Let $F(x)$ be the cdf of the exponential distribution}\\
\ \ \ \ \ \ \text{For $x \le 0$,}\ f(x) = 0 \rightarrow F(x) = \int_{- \infty}^x f(x)dx = 0\\
\ \ \ \ \ \ \text{For $x > 0$,}\ f(x) = \lambda e^{-\lambda x} \rightarrow F(x) = \int_{0}^x f(a)da = [-e^{\lambda a}]_0^x = 1 - e^{-\lambda x}\\
\ \ \ \ \ \ \therefore F(x) = \begin{cases}
1 - e^{-\lambda x}, & x > 0\\
0,\ & \text{otherwise}\\
\end{cases}\\
\ \ \ \ \ \text{(ii) The memoryless property of the exponential distribution is that the past does not affect its}\\
\ \ \ \ \ \text{future behavior. In other words, every instant can be considered as the beginning of a new}\\
\ \ \ \ \ \text{random period.}\\
\ \ \ \ \ \ \ \ \ \ \ P\{X > s+t | X > t\} = \frac{P\{X>s+t\ \text{and}\ X>t\}}{P\{X>t\}}\\
\ \ \ \ \ \ \ \ \ \ \ \text{If $X>s+t$, $X>t$ is for sure since $s>0$ in exponential distribution.}\\
\ \ \ \ \ \ \ \ \ \ \ \ \ \ = \frac{P\{X>s+t\}}{P\{X>t\}} = \frac{e^{-\lambda (s+t)}}{e^{-\lambda t}} = e^{-\lambda s} = P\{X>s\}\\
\ \ \ \ \ \ \ \ \ \ \therefore P\{X > s+t\ |\ X > t\} = P\{X>s\},\ \text{showing that $X>t$ has no effect on the probability}\\
\ \ \ \ \ \text{(iii)}\ E[X] = \int_{-\infty}^{\infty} x \lambda e^{-\lambda x} dx\\
\ \ \ \ \ \ \ \ \ \ \ = \int_{0}^{\infty} x \lambda e^{-\lambda x} dx = \frac{1}{\lambda} \int_{0}^{\infty} y e^{-y} dy\ \text{(let $y=\lambda x$, we have $dy=\lambda$)}\\
\ \ \ \ \ \ \ \ \ \ \ = \frac{1}{\lambda} [-e^{-y}-ye^{-y}]_{0}^{\infty}= \frac{1}{\lambda}\\ 
\ \ \ \ \ \ \ \ \ \ \ E[X^2] = \int_{-\infty}^{\infty} x^2 \lambda e^{-\lambda x} dx = \int_{0}^{\infty} x^2 \lambda e^{-\lambda x} dx\\
\ \ \ \ \ \ \ \ \ \ \ = \frac{1}{\lambda^2} \int_{0}^{\infty} y^2 e^{-y} dy\ \text{(let $y=\lambda x$, we have $dy=\lambda$)}\\
\ \ \ \ \ \ \ \ \ \ \ = \frac{1}{\lambda^2} [-2e^{-y}-2ye^{-y} - y^2 e^{-y}]_{0}^{\infty}= \frac{2}{\lambda^2}\\
\ \ \ \ \ \ \ \ \ \ \ \text{Var}[X] = E[X^2] - E[X]^2 = \frac{2}{\lambda^2} - \frac{1}{\lambda^2} = \frac{1}{\lambda^2}\\
\ \ \ \ \ \ \ \ \ \ \ \therefore E[X] = \frac{1}{\lambda},\ \text{Var}[X] = \frac{1}{\lambda^2}\\ \ \\
\text{(b) Let $\lambda = np$, we have $p=\frac{\lambda}{n}$}\\
\ \ \ \ \ \ \begin{equation} \begin{split}
\binom{n}{x} p^x (1-p)^{n-x} &= \binom{n}{x} (\frac{\lambda}{n})^x (1-\frac{\lambda}{n})^{n-x}\\
&= \frac{n!}{x! (n-x)!} (\frac{\lambda}{n})^x (1-\frac{\lambda}{n})^{n-x}\\
&= \frac{\lambda^x}{x!} \frac{n!}{(n-x)!} \frac{1}{n^x} (1-\frac{\lambda}{n})^{n-x}\\ \ \\
\because \frac{n!}{(n-x)!} \frac{1}{n^x} &= \frac{n(n-1)(n-2)...(n-x+1)(n-x)!}{(n-x)! n^x}\\
&= \frac{n(n-1)(n-2)...(n-x+1)}{n^x}\\
&= \frac{n}{n} \frac{n-1}{n} \frac{n-2}{n} \ ...\ \frac{n-x+1}{n}\\
&= 1(1-\frac{1}{n})(1-\frac{2}{n})\ ...\ (1-\frac{x+1}{n})\\ \ \\
\therefore \binom{n}{x} p^x (1-p)^{n-x} &= \frac{\lambda^x}{x!} (1-\frac{1}{n})(1-\frac{2}{n})\ ...\ (1-\frac{x+1}{n}) (1-\frac{\lambda}{n})^{n-x}\\
&= \frac{\lambda^x}{x!} (1-\frac{1}{n})(1-\frac{2}{n})\ ...\ (1-\frac{x+1}{n}) (1-\frac{\lambda}{n})^n (1-\frac{\lambda}{n})^{-x}\\ \ \\
\therefore \lim_{n\to\infty}\binom{n}{x} p^x (1-p)^{n-x} &= \frac{\lambda^x}{x!}\lim_{n\to\infty} [(1-\frac{1}{n})\ ...\ (1-\frac{x+1}{n})] \lim_{n\to\infty} (1-\frac{\lambda}{n})^n \lim_{n\to\infty}(1-\frac{\lambda}{n})^{-x}\\ \ \\
\because \lim_{n\to\infty} (1-\frac{\lambda}{n})^n &= e^{-\lambda}\\ \ \\
\therefore \lim_{n\to\infty}\binom{n}{x} p^x (1-p)^{n-x} &= \frac{\lambda^x}{x!}\times 1\times e^{-\lambda}\times 1 = \frac{\lambda^x}{x!}e^{-\lambda}\ \text{with $\lambda = np$}\\
\end{split} \end{equation}

\textbf{Problem 7 (Covariance and Correlation Coeffcient)}\\ \ \\
\text{(a)(i) we know that:}\\
\ \ \ \ \text{Var}(X+Y) = \text{Var}(X) + \text{Var}(Y) + 2\text{Cov}(X,Y)\\
\ \ \ \ \text{Var}(aX+b) = a^2\text{Var}(X)\\
\ \ \ \ \text{Cov}(a_1X+b_1, a_2Y+b_2) = a_1a_2\text{Cov}(X,Y)\\
\ \ \ \ \therefore \text{Var}(X+Y) = \text{Var}(X) + \text{Var}(Y) + 2\text{Cov}(X,Y) = 7\\
\ \ \ \ \ \ \ \ \ \text{Var}(2X-2Y) = \text{Var}(2X) + \text{Var}(-2Y) + 2\text{Cov}(2X,-2Y)\\
\ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ = 4\text{Var}(X) + 4\text{Var}(Y) + 2\text{Cov}(2X,-2Y)\\
\ \ \ \ \because \text{Cov}(2X,-2Y) = -4 \text{Cov}(X,Y)\\
\ \ \ \ \therefore \text{Var}(2X-2Y) = 4\text{Var}(X) + 4\text{Var}(Y) -8 \text{Cov}(X,Y) = 12\\
\ \ \ \ \therefore \text{Cov}(X,Y) = \frac{7 \times 4 - 12}{16} =1\\
\ \ \ \ \text{(ii)}\because \text{Cov}(X,Y) = 1\ \text{and}\ \text{Var}(X) = 1\\
\ \ \ \ \therefore \text{Var}(Y) = 4\\
\ \ \ \ \therefore \rho_{X,Y} = \frac{\text{Cov}(X,Y)}{\sqrt{\text{Var}(X)\text{Var}(Y)}} = \frac{1}{2}\\ \ \\
\text{(b)}\because X_1, X_2, ..., X_{10}\ \text{are uncorrelated}\\

\textbf{Problem 10 (Bayes Theorem and Conditional Probabilities)}\\ \ \\
\text{Define the following events:}\\
\text{A: autonomous vehicles have malfunctions}\\
\text{B: disengagement happens}\\
\therefore P(B|A) = 0.85,\ P(B|\bar{A}) = 0.002,\ P(A) = 0.0002\\
\therefore P(\bar{A}) = 1 - P(A) = 0.9998\\
\therefore P(B) = P(B|A)P(A) + P(B|\bar{A})P(\bar{A}) \approx 0.00217\\
\therefore P(A|B) = \frac{P(B|A)P(A)}{P(B|A)P(A) + P(B|\bar{A})P(\bar{A})} = \frac{0.85 \times 0.0002}{0.85\times 0.0002 + 0.002\times 0.9998} \approx 0.0784\\
\therefore \text{The probability that a given disengagement is due to a malfunction is about 0.0784.}

\textbf{Problem 12 (Uniform Distribution)}\\ \ \\
\text{(a)}\therefore f_X(x) = \begin{cases}
\frac{1}{10}, & 0 \le x \le 10\\
0, &\text{otherwise}
\end{cases}\\
\ \ \ \ \ \ \ \ \ f_Y(y) = \begin{cases}
\frac{1}{10}, & 0 \le y \le 10\\
0, &\text{otherwise}
\end{cases}\\
\because \text{X and Y are independent to each other}\\
\therefore f_{X,Y}(x,y) = f_X(x)f_Y(y) = \begin{cases}
\frac{1}{100}, & 0 \le x \le 10,\ 0 \le y \le 10\\
0, &\text{otherwise}
\end{cases}\\
\because P\{X=Y\} = \int_{y}^{y} \frac{1}{10}\ dx = 0\\
\therefore \text{The probability that bus A and bus B arrive at exactly the same time is 0.}\\
\\ \ \\
\text{(b)}\therefore P\{X<Y\} = \int_{0}^{10} \int_{0}^{y} \frac{1}{100}\ dxdy = \int_{0}^{10} [\frac{x}{100}]_0^y\ dy\\
\ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ = \int_{0}^{10} \frac{y}{100}\ dy = [\frac{y^2}{200}]_0^{10} = \frac{1}{2}\\
\ \ \ \ \ \therefore \text{The probability that bus A arrives earlier than bus B is $\frac{1}{2}$.}\\ \ \\
\text{(c)}\ P(Z<z) = P(x<z,y<z)\\
\because \text{X and Y are independent of each other.}\\
\therefore P(Z<z) = P(x<z)P(y<z)\\
\ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ = \int_{0}^{z} \frac{1}{10}\ dx \times \int_{0}^{z} \frac{1}{10}\ dy\\
\ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ = \begin{cases}
\frac{z^2}{100}, & 0 \le z \le 10\\
0, & \text{otherwise}\\
\end{cases}\\
\therefore f_Z(z) = \frac{d}{dz} P(Z<z) = \begin{cases}
\frac{z}{50}, & 0 \le z \le 10\\
0, & \text{otherwise}\\
\end{cases}\\
\therefore E[Z] = \int_{-\infty}^{\infty} z \cdot \frac{z}{50}\ dz = \int_{0}^{10} \frac{z^2}{50}\ dz\\
\ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ = [\frac{z^3}{150}]_0^{10} = \frac{1000}{150}\ \text{min} = 6.667\ \text{min}\\ \ \\
\text{(d)}\ P(w<W) = P(w<x,w<y)\\
\because \text{X and Y are independent of each other.}\\
\therefore P(w<W) = P(w<x)P(w<y)\\
\ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ = \int_{w}^{10} \frac{1}{10}\ dx \times \int_{w}^{10} \frac{1}{10}\ dy\\
\ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ = \begin{cases}
1-\frac{w}{5}+\frac{w^2}{100}, & 0 \le w \le 10\\
0, & \text{otherwise}\\
\end{cases}\\
\therefore P(W<w) = 1 - P(w<W) = \begin{cases}
\frac{w}{5}-\frac{w^2}{100}, & 0 \le w \le 10\\
0, & \text{otherwise}\\
\end{cases}\\
\therefore f_W(w) = \frac{d}{dw} P(W<w) = \begin{cases}
\frac{1}{5}-\frac{w}{50}, & 0 \le w \le 10\\
0, & \text{otherwise}\\
\end{cases}\\
\therefore E[W] = \int_{-\infty}^{\infty} w \cdot (\frac{1}{5}-\frac{w}{50})\ dw = \int_{0}^{10} \frac{w}{5}-\frac{w^2}{50}\ dw\\
\ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ = [\frac{w^2}{10}-\frac{w^3}{150}]_0^{10} = \frac{10}{3}\ \text{min} = 3.333\ \text{min}\\ \ \\
\text{(e)}\ \text{Define T to be the event that bus A and bus B are together at the bus stop.}\\
\text{(i) A arrives first and $x\in[0,5]$:}\ P(T_i) = \int_0^5 \int_x^{x+5} \frac{1}{100}\ dydx = \int_0^5 \frac{5}{100}\ dx = \frac{1}{4}\\
\text{(ii) A arrives first and $x\in(5,10]$:}\ P(T_{ii}) = \int_5^10 \int_x^{10} \frac{1}{100}\ dydx = \int_5^{10} \frac{1}{10}-\frac{x}{100}\ dx = \frac{1}{8}\\
\text{(iii) B arrives first and $y\in[0,5]$:}\ P(T_{iii}) = \int_0^5 \int_y^{y+5} \frac{1}{100}\ dxdy = \int_0^5 \frac{5}{100}\ dy = \frac{1}{4}\\
\text{(iv) B arrives first and $y\in(5,10]$:}\ P(T_{iv}) = \int_5^10 \int_y^{10} \frac{1}{100}\ dxdy = \int_5^{10} \frac{1}{10}-\frac{y}{100}\ dy = \frac{1}{8}\\
\therefore P(T) = \frac{1}{4} + \frac{1}{8} + \frac{1}{4} + \frac{1}{8} = \frac{3}{4}
